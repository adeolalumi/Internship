```python
'''Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
I found the following details:'''
def TeamInd():
    try:
        driver2 = Global_drive()
        url = 'https://www.bcci.tv/'
        driver2.get(url)
        time.sleep(2)

        # Click on the international dropdown
        dropdown = driver2.find_element(By.XPATH, "/html/body/header/div[3]/div[2]/ul/div[1]/a[8]")
        dropdown.click()
        time.sleep(2)

        # Extracting series names and dates
        series_names = driver2.find_elements(By.XPATH, "//div[@class='match-card__tourName ng-binding']")
        dates = driver2.find_elements(By.XPATH, "//div[@class='match-card__time ng-binding']")

        # Create a DataFrame to store the data
        data = {'Series': [], 'Date': []}

        # Iterate through the series names and dates
        for series, date in zip(series_names, dates):
            series_ = series.text.strip()
            date_ = date.text.strip()

            data['Series'].append(series_)
            data['Date'].append(date_)

        # Create a DataFrame
        df = pd.DataFrame(data)

        # Print the DataFrame
        print(df)

    except Exception as e:
        print(f"Failed to scrape Team India's fixtures: {e}")

    finally:
        driver2.quit()

TeamInd()
```

```python

def githubTRN():
    try:
        # Assuming Global_drive() is a function that returns a webdriver instance
        driver4 = Global_drive()
        url = 'https://github.com/trending'
    
        # Set up the Chrome WebDriver
        driver4.get(url)
    
        # Wait for the page to load 
        driver4.implicitly_wait(10)
        
        data3 = {'Repository Title': [], 'Repository Description': [], 'Contributors Count':[], 'Language Used':[]}
    
        # Find all repository articles
        title_ = driver4.find_elements(By.XPATH, '//h2[@class="h3 lh-condensed"]')
        descriptions_ = driver4.find_elements(By.XPATH, '//p[@class="col-9 color-fg-muted my-1 pr-4"]')
        contributors_counts = driver4.find_elements(By.XPATH, '//a[@class="Link Link--muted d-inline-block mr-3"]')
        languages_used = driver4.find_elements(By.XPATH, '//span[@class="d-inline-block ml-0 mr-3"]')
    
        # Create lists to store data
        
        for title, description, contributor, language in zip(title_, descriptions_, contributors_counts, languages_used):
            titles_ = title.text.strip()
            descriptions_ = description.text.strip()
            contributors_ = contributor.text.strip()
            languages_ = language.text.strip()
    
            data3['Repository Title'].append(titles_)
            data3['Repository Description'].append(descriptions_)
            data3['Contributors Count'].append(contributors_)
            data3['Language Used'].append(languages_)
    
        # Create a DataFrame
        df = pd.DataFrame(data3)
    
        # Print the DataFrame
        print(df)

    except Exception as e:
        print(f"Failed to scrape GitHub Trending: {e}")

    finally:
        driver4.quit()

if __name__ == "__main__":
    githubTRN()
```
